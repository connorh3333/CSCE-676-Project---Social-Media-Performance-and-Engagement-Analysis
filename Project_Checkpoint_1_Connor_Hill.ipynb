#**(D) Exploratory Data Analysis**

```python
from google.colab import files
from google.colab import drive
import os
import shutil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')

print(os.listdir('/content/drive/MyDrive/CSCE676/data'))


df = pd.read_csv('/content/drive/MyDrive/CSCE676/data/social_media_performance.csv')


# print the shape of the dataset
print("SHAPE: ", df.shape)
print("\n#################################\n")

# print the columns of the datset
print("COLUMNS: ", df.columns)
print("\n#################################\n")

# Check to see if there are any missing values
print("MISSING VALUES: ", df.isnull().sum())
print("\n#################################\n")


# Display the first 10 rows of the datset to give a general overview of the dataset's structure and datatypes
print("FIRST 10 ROWS OF THE DATASET:")
display(df.head(10))
print("\n#################################\n")


# Check for duplicate rows

duplicates = df.duplicated().sum()

print("NUMBER OF DUPLICATE ROWS IN THE DATASET: ", duplicates, "\n")
print("\n#################################\n")

# Display descriptive statistics for numeric and categorical data in the set
print("DESCRIPTIVE STATISTICS FOR NUMERIC DATA:")


display(df.describe(include=[np.number]).T)
print("\n#################################\n")

print("\nDESCRIPTIVE STATISTICS FOR THE CATEGORICAL DATA:")

display(df.describe(include=["object"]).T if len(df.select_dtypes(include=["object"]).columns) else pd.DataFrame({"message": ["No object/text columns found"]}))
print("\n#################################\n")

# Perform Categorical Data Analysis

print("COLUMN BY COLUMN CATEGORICAL DATA ANALYSIS")

category_cols = df.select_dtypes(include=["object"]).columns.tolist()

for col in category_cols:
  print(f"\n--- {col} ---")
  print("Unique values:", df[col].nunique())
  display(df[col].value_counts(dropna=False).head(10))

print("\n#################################\n")

# Find the numeric distributions and outliers

print("NUMERIC DISTRIBUTIONS AND OUTLIERS")

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

cols_to_plot = num_cols[:12]
df[cols_to_plot].hist(bins=30, figsize=(14,10))
plt.suptitle("Numeric Distributions (first 12 numeric columns)")
plt.tight_layout()
plt.show()

cols_box = num_cols[:8]
plt.figure(figsize=(14,5))
df[cols_box].plot(kind="box")
plt.title("Outlier Check (for first 8 numeric columns)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Outlier counts via IQR

print("\nOutlier counts (IQR method):")
outlier_rows = {}
for col in num_cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)

  iqr = q3 - q1

  if pd.isna(iqr) or iqr == 0:
    outlier_rows[col] = 0
    continue

  low = q1 - 1.5 * iqr
  high = q3 + 1.5 * iqr

  outlier_rows[col] = int(((df[col] < low) | (df[col] > high)).sum())

outlier_df = pd.DataFrame.from_dict(outlier_rows, orient="index", columns=["outlier_count"])
outlier_df = outlier_df.sort_values("outlier_count", ascending=False)
display(outlier_df.head(15))

print("\n#################################\n")
```

**Algorithm Explanation:**
- **IQR Method:**
  - I used the IQR method to find the number of outliers in the data because social media data is usually heavily skewed and the IQR method is good for analyzing skewed data because the method itself is not affected by extreme values

- **Hypothesis:**

Sharp spikes in likes, views, comments and shares suggest that there are a very select few posts that people enjoy, but the more even distribution of engagement rate suggests that the majority of posts have a decent amount of engagement from their viewers.
