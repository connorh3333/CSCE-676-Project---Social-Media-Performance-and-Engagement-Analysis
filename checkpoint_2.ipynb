#**1. Project Scope:**
- **Dataset:** Social Media Performance and Engagement Data from the Kaggle website and created by S.V.Thejaswini
- **Course Technique:** Association Rule Mining
- **External Technique:** Time-series Forecasting

#**2. Research Question Definition:**
- **1.) Questions Relevant to Course Techniques:**
  -  **What are the associations between content types and content topics?**
    - i.e.
      - Do topics about lifestyle tend to be in video format?
      - Do posts relating to food tend to be in image format?
    - **Data Mining Task Type:**
      - Association Rule Mining
    - **Relevant Algorithm(s):**
      - SON Algorithm
    - **Evaluation Criteria:**
      - Confidence
        - Support (needed to calculate confidence)
      - Interest
      - Lift
  - **2.) What is the association between a post's sentiment and their number of likes?**
    - i.e.
      - Does a post with a more positve sentiment tend to have more likes than a post that is more negative?
    - **Data Mining Task Type:**
      - Association Rule Mining
    - **Relevant Algorithm(s):**
      - SON Algorithm
    - **Evaluation Criteria:**
      - Confidence
        - Support (needed to calculate confidence)
      - Interest
      - Lift
- **Question Relevant to External Technique:**
  - **3.) What time periods, for the next year, will there be the largest influxes in post count in social media?**
    - i.e.
      - Will there be more social media postings in next year's Summer season, Fall?
        - If during a specific season, which months will be the most popular for social media posting
    - **Data Mining Task Type:**
      - Time-series Forecasting
    - **Relevant Algorithm(s):**
      - ARIMA (Auto Regressive Integrated Moving Average) model
      - SARIMA (Seasonal Auto Regressive Integrated Moving Average) model
    - **Evaluation Criteria:**
      - RMSE (Root Mean Squared Error)


#**3. Motivation and Feasibility:**
- **Motivation:**
  - 1.) My EDA from the previous project checkpoint states that the top content topic is "Lifestyle" and the most used content type is "video" but it dos not present the correlation between the two
    - i.e.
      - What percentages of all of the videos are about "lifestyle"?
  - 2.) The EDA's sentiment score distribution is left skewed which implies that the number of "postive posts" greatly outweighs the number of posts with more negative sentiment scores
  - 3.) The EDA had shown that the top 3 post datetimes were "2025-09-08 10:37:00," "2025-11-03 07:49:00," and "2025-10-25 19:01:00" implying that social media users more frequently post during the Fall season
- **Non-triviality:**
  - Association Rule Mining describes relationship between items, not ranges of values ([x1,xn] --> [y1, yn])
  - Temporal ordering and clustered surges make RQ #3 more challenging than a standard non-temporal mining task
- **Feasability:**
  - All algorithms, SON, ARIMA, and SARIMA, are learnable and implementable
  - Data parameters with large ranges like sentiment scores, number of likes, large influxes in post counts, can be converted into lists of smaller intervals: [x1, xn] = {[x1,x100], [x101, x500], [x501, x1000], ..., [xn-num, xn]}
- **Risks:**
  - Large computation time due to size of dataset
  - Large costs in memory

#**4. Methodological Planning:**
- **Course Algorithm:**
  - SON algorithm
- **External Algorithms:**
  - ARIMA Algorithm
  - SARIMA Algorithm
- **Evaluation:**
  - Objective measures of accuracy and correlation for association rules:
    - Confidence
    - Support
    - Interest
    - Lift
  - Accuracy score for time-series forecasting algorithm's predictions in the form of an error metric on the same scale as the original data
    - Root Mean Squared Error (RMSE)
- **Baselines:**
  - Apriori Algorithm
    - Benchmark to compare to SON
  - AR (Auto Regressive) model
    - Benchmark used to compare ARIMA and SARIMA
